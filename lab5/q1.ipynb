```python
n=int(input())
m=int(input())
k=int(input())
##preparing data
obs_kids_weights = np.zeros(int(n/m))
obs_kids_heights = np.zeros(int(n/m))
obs_adults_weights = np.zeros(int(n/m))
obs_adults_heights = np.zeros(int(n/m))
for i in range(int(n/m)):
    obs_kids_weights[i]=np.random.uniform(40, 50)
    obs_kids_heights[i]=np.random.uniform(4, 5)
    obs_adults_weights[i]=np.random.uniform(60, 70)
    obs_adults_heights[i]=np.random.uniform(5, 6)
obs_weights=np.append(obs_kids_weights, obs_adults_weights)
obs_heights=np.append(obs_kids_heights, obs_adults_heights)
obs=[]
for i in range(len(obs_weights)):
    obs.append([obs_heights[i], obs_weights[i]])
obs=np.array(obs)
#plotting data
plt.scatter(obs_heights, obs_weights, color='cyan', marker='.')
plt.show()
##algorithm
def k_means_cluster_gen(centres, obs, track_centres):
    while(1):
        track_centres=np.copy(centres)
        color = ['red', 'blue']
        centre=np.zeros(len(centres))
        for k_ in range(len(centres)):
            S=np.array([])
            for i in range(len(obs)):
                count=0
                for k1 in range(len(centres)):
                    if(k1!=k_):
                        if(dist(obs[i], centres[k_])<dist(obs[i], centres[k1])):
                            count+=1
                if(count==len(centres)-1):
                    S=S.tolist()
                    S.append(obs[i])
                    S=np.array(S)
            if(len(S)>0):
                centre=centre.tolist()
                centre[k_]=update_centre(S)
                centre=np.array(centre)
            else:
                centre=centre.tolist()
                centre[k_]=centres[k_]
                centre=np.array(centre)
            plt.title('2-d cluster')
            plt.scatter(S[:,1], S[:,0], marker='.', color=color[k_])
        centres=centres.tolist()
        centres=copy.deepcopy(centre)
        centres=np.array(centres)
        plt.scatter(centres[:,1], centres[:,0], marker='*', color='black')
        plt.show()
        print(centres)
        if(np.array_equal(track_centres, centres)):
            break
    return centres
def dist(o, c):
    return ((o[0]-c[0])**2)+((o[1]-c[1])**2)
def update_centre(comprehension):
    l=0
    x=0
    y=0
    for ele in comprehension:
        x+=ele[0]
        y+=ele[1]
        l+=1
    return (x/l, y/l)
#function calling
g_centres=[[4.5, 60], [5.5, 65]]
g_centres=np.array(g_centres)
for i in range(4):
#     print("GUESS CENTRES==", g_centres[i])
    track_centres = np.array([])
    final_centres=k_means_cluster_gen(g_centres, obs, track_centres)
```
