```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def plotter_f1():
    w1=np.linspace(-10, 10, 15)
    w2=np.linspace(-10, 10, 15)
    W1, W2=np.meshgrid(w1, w2)
    f1=((W1**2)/2)+((W2**2)/2)
    plt.contour(w1, w2, f1)
    plt.title('contour of f1')
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.grid(True)
    x, y=np.gradient(f1)
    plt.quiver(w1, w2, x, y, color='k')
def plotter_f2():
    w1=np.linspace(-10, 10, 15)
    w2=np.linspace(-10, 10, 15)
    W1, W2=np.meshgrid(w1, w2)
    f2=((W1**2)*5)+((W2**2)/2)
    plt.contour(w1, w2, f2)
    plt.title('contour of f2')
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.grid(True)
    x, y=np.gradient(f2)
    plt.quiver(w1, w2, x, y, color='k')
def plotter_f3():
    w1=np.linspace(-10, 10, 15)
    w2=np.linspace(-10, 10, 15)
    W1, W2=np.meshgrid(w1, w2)
    f3=((W1**2)/2)+((W2**2)*5)
    plt.contour(w1, w2, f3)
    plt.title('contour of f3')
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.grid(True)
    x, y=np.gradient(f3)
    plt.quiver(w1, w2, x, y, color='k')
def plotter_f4():
    w1=np.linspace(-10, 10, 15)
    w2=np.linspace(-10, 10, 15)
    W1, W2=np.meshgrid(w1, w2)
    f4=((W1**2)/2)+((W2**2)/2)+(5*W1)-(3*W2)-2
    plt.contour(w1, w2, f4)
    plt.title('contour of f4')
    plt.xlabel('w1')
    plt.ylabel('w2')
    plt.grid(True)
    x, y=np.gradient(f4)
    plt.quiver(w1, w2, x, y, color='k')
plotter_f1()
plt.show()
plotter_f2()
plt.show()
plotter_f3()
plt.show()
plotter_f4()
plt.show()
```

```python
#######f1_converging
def diff_wrt_w1(w1):
    return w1
def diff_wrt_w2(w2):
    return w2
def gradient_descent_2d(w1, w2, alpha):
    w1_track, w2_track=0, 0
    itr=0
    plotter_f1()
    while(w1_track!=w1 and w2_track!=w2):
        w1_track, w2_track=w1, w2
        w1_dash, w2_dash=diff_wrt_w1(w1), diff_wrt_w2(w2)
        w1, w2=w1-(alpha*w1_dash), w2-(alpha*w2_dash)
        plt.scatter(w1, w2, color='red', marker='.')
        itr+=1
        print(w1, w2)
    plt.scatter(w1, w2, color='chocolate', marker='*')
    plt.show()
    return w1, w2, itr
w1=3
w2=2
alpha=0.9
w1, w2, itr=gradient_descent_2d(w1, w2, alpha)
```
