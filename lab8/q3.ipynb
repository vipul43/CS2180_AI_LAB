```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
#reading data from text file
#data are points in 2D plane
f=open("logistic", "r")
text=f.read().split()
x1=[]
x2=[]
y=[]
n=20
for i in range(1, 21):
    x1.append(float(text[3*i]))
    x2.append(float(text[3*i+1]))
    y.append(float(text[3*i+2]))
```

```python
def dataset_plotter():
    for i in range(n):
        if(y[i]==+1):
            plt.scatter(x1[i], x2[i], color='red', marker='.')
        else:
            plt.scatter(x1[i], x2[i], color='blue', marker='.')
    plt.show()
dataset_plotter()
```

```python
def sigmoid(z):
    return float(1/(1+exp(-z)))
def delta(w, n):
    s=np.array([0, 0 ,0])
    for i in range(n):
        z=y[i]*(np.dot(w, [x2[i], x1[i], 1]))
        s=s+((1-sigmoid(z))*(np.array([x2[i], x1[i], 1])*y[i]))
    return s
def loss_function(w):
    loss=0
    for i in range(n):
        z=y[i]*(np.dot(w, [x2[i], x1[i], 1]))
        loss+=(-log(sigmoid(z)))
    return loss
def logistic_regression(w, alpha, TOL):
    itr=0
    loss=999999
    while(itr!=100):
        loss=loss_function(w)
        print('loss=', loss)
        w=w+(alpha*delta(w, n))
        itr+=1
    return w, itr
#####calling functions######
n=20
w_initial=np.array([1, 1, 1])
alpha=0.1
TOL=0.01
w, itr=logistic_regression(w_initial, alpha, TOL)
print('w=', w, 'itr=', itr)

```
