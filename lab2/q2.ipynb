```python
import numpy as np
def environment(x_size, y_size):
    return np.zeros((x_size, y_size))
```

```python
def add_dirt(room):
    x=np.random.randint(0, room.shape[0])
    y=np.random.randint(0, room.shape[1])
    room[x][y]+=1
    return x, y
```

```python
def action(obs_x, obs_y, room):
    a=np.random.randint(0, 5)
    r=0
    if(a==0):
        action='pickdirt'
        if(room[obs_x][obs_y]!=0):
            r = room[obs_x][obs_y]
            room[obs_x][obs_y]=0
        else:
            r = get_reward(action)
    elif(a==1):
        action='up'
        if(obs_x!=0):
            obs_x-=1
        else:
            r = get_reward(action)
    elif(a==2):
        action='down'
        if(obs_x!=room.shape[0]-1):
            obs_x+=1
        else:
            r = get_reward(action)
    elif(a==3):
        action='left'
        if(obs_y!=0):
            obs_y-=1
        else:
            r = get_reward(action)
    else:
        action='right'
        if(obs_y!=room.shape[1]-1):
            obs_y+=1
        else:
            r = get_reward(action)
    print("action of agent: ", action)
    return obs_x, obs_y, r
```

```python
def get_reward(action):
    if(action=='pickdirt'):
        reward=-1
    else:
        reward=-10
    return reward
```
