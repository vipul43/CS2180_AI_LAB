```python
import numpy as np
A = [1, 2, 3, 4, 6]
p_out_min = [0.01, 0.02, 0.03, 0.1, 0.3]
p_out_max = [0.1, 0.2, 0.3, 0.5, 0.7]
p_run_min = 0.5
p_run_max = 0.8
```

```python
def environment(s, runs, a):
#     a=action(s)
    p_out = p_out_max[a]+((p_out_min[a]-p_out_max[a])*((s[1]-1)/9))
    p_run = p_run_min + ((p_run_max-p_run_min)*((s[1]-1)/9))
    s[0]-=1
    if(a==0):
        o=np.random.rand()
        if(o<p_out):
            #OUT!!!
            runs=0
            s[1]-=1
        else:
            #NOT OUT
            r=np.random.rand()
            if(r<p_run):
                #SCORES
                runs=A[0]
            else:
                #DOES NOT SCORE
                runs = 0
    elif(a==1):
        o=np.random.rand()
        if(o<p_out):
            #OUT!!!
            runs=0
            s[1]-=1
        else:
            #NOT OUT
            r=np.random.rand()
            if(r<p_run):
                #SCORES
                runs=A[1]
            else:
                #DOES NOT SCORE
                runs = 0
    elif(a==2):
        o=np.random.rand()
        if(o<p_out):
            #OUT!!!
            runs=0
            s[1]-=1
        else:
            #NOT OUT
            r=np.random.rand()
            if(r<p_run):
                #SCORESki
                runs=A[2]
            else:
                #DOES NOT SCORE
                runs = 0
    elif(a==3):
        o=np.random.rand()
        if(o<p_out):
            #OUT!!!
            runs=0
            s[1]-=1
        else:
            #NOT OUT
            r=np.random.rand()
            if(r<p_run):
                #SCORES
                runs=A[3]
            else:
                #DOES NOT SCORE
                runs = 0
    elif(a==6):
        o=np.random.rand()
        if(o<p_out):
            #OUT!!!
            runs=0
            s[1]-=1
        else:
            #NOT OUT
            r=np.random.rand()
            if(r<p_run):
                #SCORES
                runs=A[4]
            else:
                #DOES NOT SCORE
                runs = 0
    return s, runs
def action(s):
    return np.random.randint(0, 5)
```

```python
a=0

s = [300, 10]
runs=0
total_score =0
print(s, 0)
balls_played_by_one=0
balls_played_by_last=0
runs_scored_by_one=0
runs_scored_by_last=0
while(s[0]!=0 and s[1]!=1):
    s, runs = environment(s, runs, a)
    total_score = runs+total_score
    if(s[1]==10):
        balls_played_by_one+=1
        runs_scored_by_one+=runs
    if(s[1]==1):
        balls_played_by_last+=1
        runs_scored_by_last+=runs
    print(s, runs)
print(total_score)
print("average balls played by one is: ", balls_played_by_one)
print("average balls played by last is: ", balls_played_by_last)
print("average runs scored by one is: ", runs_scored_by_one)
print("average runs scored by last is: ", runs_scored_by_last)
```
