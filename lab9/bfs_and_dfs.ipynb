```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10
graph=[]
for i in range(n*n):
    u=[]
    graph.append(u)
graph
k=0
obstacles=[[3, 3], [4, 3], [5, 3], [3, 4], [5, 4], [3, 5], [5, 5]]
for i in range(n):
    for j in range(n):
        if(j+1<=n-1 and [j+1, i] not in obstacles):
            graph[k].append([j+1, i])
        if(j-1>=0 and [j-1, i] not in obstacles):
            graph[k].append([j-1, i])
        if(i+1<=n-1 and [j, i+1] not in obstacles):
            graph[k].append([j, i+1])
        if(i-1>=0 and [j, i-1] not in obstacles):
            graph[k].append([j, i-1])
        ##############################
#         if(i+1<=n-1 and j+1<=n-1):
#             graph[k].append([j+1, i+1])
#         if(i-1>=0 and j-1>=0):
#             graph[k].append([j-1, i-1])
#         if(i-1>=0 and j+1<=n-1):
#             graph[k].append([j+1, i-1])
#         if(i+1<=n-1 and j-1>=0):
#             graph[k].append([j-1, i+1])
        k+=1
######obstacle_inducer#############
for i in range(n):
    for j in range(n):
        if([j, i] in obstacles):
            graph[j+(n*i)]=[]
for i in range(n*n):
    print("i=", i, "   ", graph[i]) 
grid=np.zeros((n, n))
```

```python
def plot_path(path):
    for point in path:
        plt.scatter(point[0], point[1])
    plt.grid(True)
    plt.show()
def index(li):
    return int(li[0]+n*(li[1]))
def bfs(graph, n, s, d):
    visited=[False]*(n*n)
    queue=[]
    path=[]
    queue.append(s)
    visited[index(s)]=True
    while queue:
        for point in obstacles:
            plt.scatter(point[0], point[1], color='black')
        s=queue.pop(0)
        path.append(s)
        plot_path(path)
        if(s==d):
            break
        for v in graph[index(s)]:
            if(visited[index(v)]==False):
                queue.append(v)
                visited[index(v)]=True
    return path

#########################
path=[]
s=[0, 0]
d=[4, 4]
path=bfs(graph, n, s, d)
print("FINAL PATH=", path)
```
