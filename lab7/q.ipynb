```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def func(a, b, c, w):
    return ((a*(w**2))/2)+(b*w)+c
def diff_wrt_w(a, b, c, w):
    return (a*w)+b
def gradient_descent_1d(a, b, c, w, alpha, plot_list):
    w_track=0
    itr=0
    while(w!=w_track):
        w_track=w
        w_minus_w_star=w+(b/a)
        plot_list.append(w_minus_w_star)
        plt.scatter(w, func(a, b, c, w), color='r', marker='.')
        w=w-(alpha*diff_wrt_w(a, b, c, w))
        itr+=1
    return w, plot_list, itr
        
    
```

```python
a_list=[1, 0.1, 10, 1, 1, 0.1]
b_list=[0, 0, 0, 0, 1, -1]
c_list=[0, 0, 0, 10, 1, -1]
alpha_list=[1.1, 12, 0.10001, 1.4, 1.5, 15]
for i in range(6):
    a=a_list[i]
    b=b_list[i]
    c=c_list[i]
    print('a, b, c=', a, b, c)
    alpha=alpha_list[i]
    x=np.linspace(-20, 20, 300)
    y=func(a, b, c, x)
    plt.plot(x, y, color='yellow')
    w=5
    plot_list=[]
    w_star, plot_list, itr=gradient_descent_1d(a, b, c, w, alpha, plot_list)
    print(w_star, itr)
    plt.scatter(w_star, func(a, b, c, w_star), color='k', marker='*')
    plt.show()
    t=np.arange(0, itr, 1)
    plt.plot(t, plot_list, color='cyan')
    plt.show()
```
