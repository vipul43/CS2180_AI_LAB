```python
import numpy as np
from math import *
import numpy as np
import matplotlib.pyplot as plt
from math import *
def environment_village(p_kid):
    person=np.random.rand()
    if(person<p_kid):
        height=np.random.normal(u11, var11)
        weight=np.random.normal(u12, var12)
        heights_giv_kid.append(height)
        weights_giv_kid.append(weight)
        person=0
    else:
        height=np.random.normal(u21, var21)
        weight=np.random.normal(u22, var22)
        heights_giv_adult.append(height)
        weights_giv_adult.append(weight)
        person=1
    return person, height, weight

def agent(p_kid, height, weight):
    if(p_kid*p_x_giv_(u11, var11, u12, var12, height, weight)>(1-p_kid)*p_x_giv_(u21, var21, u22, var22, height, weight)):
        return 0
    else:
        return 1
def p_x_giv_(u1, var1, u2, var2, x, z):
    y=(1/((2*pi)**0.5)*var1)*exp(-(1/2)*(((x-u1)/var1))**2)*(1/((2*pi)**0.5)*var2)*exp(-(1/2)*(((z-u2)/var2))**2)
    return y

p_kid=float(input())
persons=[]
heights_total=[]
weights_total=[]
heights_giv_kid=[]
heights_giv_adult=[]
weights_giv_kid=[]
weights_giv_adult=[]


loss=0
u11=3
u12=4
u21=3
u22=4
var11=0.3
var12=0.2
var21=0.3
var22=0.4

for i in range(1000):
    person, height, weight=environment_village(p_kid)
    persons.append(person)
    heights_total.append(height)
    weights_total.append(weight)
    decision=agent(p_kid, height, weight)
    if(decision==person):
        loss+=0
    else:
        loss+=1
print(loss/1000)
    
plt.hist(persons)
plt.title("kid_or_adult")
plt.show()
plt.hist(heights_total)
plt.title("heights")
plt.show()
plt.hist(weights_total)
plt.title("weights")
plt.show()
plt.hist(heights_giv_kid)
plt.title("heights_giv_kids")
plt.show()
plt.hist(heights_giv_adult)
plt.title("heights_giv_adults")
plt.show()
plt.hist(weights_giv_kid)
plt.title("weights_giv_kids")
plt.show()
plt.hist(weights_giv_adult)
plt.title("weights_giv_adults")
plt.show()
```
